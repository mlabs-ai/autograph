```python
import json

with open("../data/clusters/P171.json", "r") as file:
    obj = json.load(file)
```

```python
important_ids = set()
for v in obj.values():
    important_ids.update(v)
```

```python
d = {}
with open("../data/wikidata/latest-all.json", 'r') as file:
    for line in file:
        try:
            item = json.loads(line[:-2])
        except Exception as e:
            print(e)
            continue
        key = item["id"]
        if key in important_ids:
            labels = item["labels"]
            if "en" in labels:
                label = labels["en"]["value"]
            elif len(labels) > 0:
                label = list(labels.values())[0]["value"]
            else:
                label = key
            d[key] = label

            if len(d) % 1000 == 0:
                print(f"{len(d)}/{len(important_ids)}")
```

```python
obj = {i: [d[v] if v in d else v for v in vs] for i, vs in obj.items()}
```

```python
with open("../data/clusters/P171_renamed.json", "w") as file:
    json.dump(obj, file, indent=4)
```
